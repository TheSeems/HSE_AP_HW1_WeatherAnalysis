```python
import pandas as pd
import numpy as np
import multiprocessing as mp
```

```python
# Реальные средние температуры (примерные данные) для городов по сезонам
seasonal_temperatures = {
    "New York": {"winter": 0, "spring": 10, "summer": 25, "autumn": 15},
    "London": {"winter": 5, "spring": 11, "summer": 18, "autumn": 12},
    "Paris": {"winter": 4, "spring": 12, "summer": 20, "autumn": 13},
    "Tokyo": {"winter": 6, "spring": 15, "summer": 27, "autumn": 18},
    "Moscow": {"winter": -10, "spring": 5, "summer": 18, "autumn": 8},
    "Sydney": {"winter": 12, "spring": 18, "summer": 25, "autumn": 20},
    "Berlin": {"winter": 0, "spring": 10, "summer": 20, "autumn": 11},
    "Beijing": {"winter": -2, "spring": 13, "summer": 27, "autumn": 16},
    "Rio de Janeiro": {"winter": 20, "spring": 25, "summer": 30, "autumn": 25},
    "Dubai": {"winter": 20, "spring": 30, "summer": 40, "autumn": 30},
    "Los Angeles": {"winter": 15, "spring": 18, "summer": 25, "autumn": 20},
    "Singapore": {"winter": 27, "spring": 28, "summer": 28, "autumn": 27},
    "Mumbai": {"winter": 25, "spring": 30, "summer": 35, "autumn": 30},
    "Cairo": {"winter": 15, "spring": 25, "summer": 35, "autumn": 25},
    "Mexico City": {"winter": 12, "spring": 18, "summer": 20, "autumn": 15},
}

# Сопоставление месяцев с сезонами
month_to_season = {12: "winter", 1: "winter", 2: "winter",
                   3: "spring", 4: "spring", 5: "spring",
                   6: "summer", 7: "summer", 8: "summer",
                   9: "autumn", 10: "autumn", 11: "autumn"}

# Генерация данных о температуре
def generate_realistic_temperature_data(cities, num_years=10):
    dates = pd.date_range(start="2010-01-01", periods=365 * num_years, freq="D")
    data = []

    for city in cities:
        for date in dates:
            season = month_to_season[date.month]
            mean_temp = seasonal_temperatures[city][season]
            # Добавляем случайное отклонение
            temperature = np.random.normal(loc=mean_temp, scale=5)
            data.append({"city": city, "timestamp": date, "temperature": temperature})

    df = pd.DataFrame(data)
    df['season'] = df['timestamp'].dt.month.map(lambda x: month_to_season[x])
    return df

# Генерация данных
data = generate_realistic_temperature_data(list(seasonal_temperatures.keys()))
data.to_csv('temperature_data.csv', index=False)
```

```python
data = pd.read_csv('temperature_data.csv')
data['timestamp'] = pd.to_datetime(data['timestamp'])
data
```

(Прямолинейно) Вычислим среднее и найдем аномалии

```python
def process_data(data):
  data['temp_m_avg'] = data.groupby('city')['temperature'].transform(lambda x: x.rolling(window=30, center=True).apply(np.nanmedian))
  data['temp_m_std'] = data.groupby('city')['temperature'].transform(lambda x: x.rolling(window=30, center=True).apply(np.nanstd))
  data[['temp_m_avg', 'temp_m_std']][data['temp_m_avg'].notna()]
  data['anomaly'] = (data['temperature'] < data['temp_m_avg'] - 2 * data['temp_m_std']) | (data['temperature'] > data['temp_m_avg'] + 2 * data['temp_m_std'])
  return data
```

```python
%%time

copied_data = data.copy(deep=True)
linear_results = process_data(copied_data)
```

```python
linear_results[linear_results['anomaly'] == True]
```

```python
%%time

copied_data = data.copy(deep=True)
city_groups = data.groupby('city')
city_dataframes = [group for _, group in city_groups]
with mp.Pool(processes=mp.cpu_count()) as pool:
    results = pool.map(process_data, city_dataframes)

parallel_results = pd.concat(results)
```

```python
parallel_results[parallel_results['anomaly'] == True]
```

Получим текущую температуру для `Beijing`

```python
current_city = "Beijing"
```

```python
def get_current_temperature(city):
    params = {
        "q": city,
        "appid": OPENWEATHERMAP_API_KEY,
        "units": "metric",
    }
    response = requests.get(OPENWEATHERMAP_WEATHER_API_URL, params=params)
    response.raise_for_status()
    data = response.json()
    return data["main"]["temp"]
```

Посмотрим насколько температура соответствует норме по имеющимся данным

```python
month_to_season = {
    12: "winter", 1: "winter", 2: "winter",
    3: "spring", 4: "spring", 5: "spring",
    6: "summer", 7: "summer", 8: "summer",
    9: "autumn", 10: "autumn", 11: "autumn"
}
```

Проагрегируем по временам года

```python
seasonal_stats = data.groupby(['city', 'season'])['temperature'].agg(['mean', 'std'])
seasonal_stats = seasonal_stats.rename(columns={'mean': 'seasonal_mean', 'std': 'seasonal_std'})
data = data.merge(seasonal_stats, on=['city', 'season'], how='left')
current_city_data = data[data['city'] == current_city]
current_city_data
```

```python
def is_temperature_normal(current_city, current_city_temp, season):
  seasonal_mean = data[(data['city'] == current_city) & (data['season'] == season)]['seasonal_mean'].values[0]
  seasonal_std = data[(data['city'] == current_city) & (data['season'] == season)]['seasonal_std'].values[0]
  return seasonal_mean - 2 * seasonal_std <= current_city_temp <= seasonal_mean + 2 * seasonal_std
```

```python
'Fetched temperature can be considered \'normal\': ', is_temperature_normal(current_city, current_city_temp, season)
```

Пробуем разные города (асинхронно)

```python
async def get_current_temperature_async(city, api_key):
    params = {
        "q": city,
        "appid": OPENWEATHERMAP_API_KEY,
        "units": "metric",
    }
    async with aiohttp.ClientSession() as session:
        async with session.get(OPENWEATHERMAP_WEATHER_API_URL, params=params) as response:
            data = await response.json()
            return data["main"]["temp"]
```

```python
unique_cities = data['city'].unique()
tasks = [get_current_temperature_async(city, OPENWEATHERMAP_API_KEY) for city in unique_cities]
results = await asyncio.gather(*tasks)
for city, temp in zip(unique_cities, results):
  print(city, ":", is_temperature_normal(city, temp, season))
```
